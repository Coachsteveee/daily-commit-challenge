```python
import seaborn as sns
%matplotlib inline
import numpy as np
import matplotlib as plt
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.displot(tips['total_bill'], bins=12,kde=True,)
```

```python
 sns.jointplot(x='tip',y='total_bill',data=tips,kind='scatter')
```

```python
sns.pairplot(tips, hue='sex', palette='coolwarm')
```

```python
sns.rugplot(tips['total_bill'])
```

MATRIX PLOTS

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
fc = flights.corr()
```

```python
sns.heatmap(fc, annot=True, cmap='magma')
```

```python
tc = tips.corr()
sns.heatmap(tc, cmap='pink_r')
```

```python
fp = flights.pivot_table(index='month',columns='year',values='passengers')
```

```python
sns.heatmap(fp, cmap='coolwarm',linecolor='grey',linewidths=2)
```

```python
fp
```

```python
sns.clustermap(fp, cmap= 'coolwarm', standard_scale=1)
```

GRIDS

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
g=sns.PairGrid(iris)
g.map_diag(sns.distplot, color='pink')
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
```

```python
t = sns.FacetGrid(tips,col='time',row='smoker')
t.map(sns.distplot, 'total_bill')
```

REGRESSION PLOTS

```python
sns.lmplot(x='total_bill',y='tip',data=tips, hue='smoker',markers=['o','v'], scatter_kws={'s':100})
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex',hue='time',palette='viridis')
```
